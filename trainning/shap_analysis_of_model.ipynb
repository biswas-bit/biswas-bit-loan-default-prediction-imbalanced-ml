```python
import joblib
import shap
```

```python
model = joblib.load("/content/loan_default_model.pkl")
X_sample = joblib.load("/content/X_sample.pkl")
```

```python
print(X_sample.shape)
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer(X_sample)
print("Shap values shape:", shap_values.shape)
```

```python
shap.summary_plot(shap_values.values,X_sample)
```

```python
shap.summary_plot(shap_values.values, X_sample,plot_type='bar')
```

```python
# prefeature contribution  to this prediction
sample_index = 0
shap.plots.waterfall(shap_values[sample_index])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
shap.summary_plot(shap_values.values,X_sample)
plt.savefig('shap_summary.png',bbox_inches='tight')
```

```python
plt.figure(figsize=(10,6))
shap.summary_plot(shap_values.values, X_sample, plot_type="bar", show=False)
plt.savefig("shap_bar.png", bbox_inches="tight")
```

```python
plt.figure(figsize=(10,6))
sample_index = 0
shap.plots.waterfall(shap_values[sample_index])
plt.savefig("shap_waterfall.png", bbox_inches="tight")
```
